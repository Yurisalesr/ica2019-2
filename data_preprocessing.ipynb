````python
# Importing dataset:
data = pd.read_excel('mos_dataset.xlsx')
# Visualizing data:
display(data.head())
# Display data info:
data.info()
````

Como observado acima, faz-se necessário o tratamento das variáveis não numéricas, introduzindo $\textit{dummy variables}$ para representar os preditores categóricos `Call Test Result` e `Call Test Technology`.

````python
print(f'''Categorical data:\n\n{data['Call Test Result'].value_counts()}
\n{data['Call Test Technology'].value_counts()}''')
````

````python
# Transforming categorical data:
# in 'Call Test Result': 'SUCCESS' and 'FAILURE - DROP CALL' to 1 and 0:
data = data.replace(to_replace={'Call Test Result': 'SUCCESS'}, value=1)
data = data.replace(to_replace={'Call Test Result': 'FAILURE - DROP CALL'}, value=0)
print(data['Call Test Result'].value_counts(),'\n')

# in 'Call Test Technology': 'LTE', 'UMTS', and 'GSM' to 2, 1 and 0:
data = data.replace(to_replace={'Call Test Technology': 'LTE'}, value=2)
data = data.replace(to_replace={'Call Test Technology': 'UMTS'}, value=1)
data = data.replace(to_replace={'Call Test Technology': 'GSM'}, value=0)
print(data['Call Test Technology'].value_counts(),'\n')
````

Observando os dados após a transformação, temos:

````python
data.head(6)
````

````python
# Replacing negative values in 'Speed (m/s)':
data = data.replace(to_replace={'Speed (m/s)': -1}, value=0)
````

Para simplificar o trabalho de classificação dos dados, faz-se necessário mapear os valores de ```'MOS'``` em 
níveis discretos. 

De acordo com a ITU, os níveis de Mean Opinion Score são divididos em 5 categorias de julgamento:

| Rating | User satisfaction |$\text{MOS}_{CQE}$|
|--------|-------------------|-----------|
|  Bad (1)     | Nearly all users dissatisfied | 2.58 |
|Poor (2)| Many users dissatisfied  |3.10|
|Fair (3)|Some users dissatisfied| 3.60 |
|Good (4)| Satisfied  |4.03|
|Excelent (5)| Very satisfied  |4.34|

onde $\text{MOS}_{CQE}$ se refere à qualidade de convesação estimada.   
Reduzindo o numero de classes de a cordo com a recomendação, fazemos:

````python
# Transforming the 'MOS' column values into 5 classes (1-5):
data.loc[(data['MOS']>=1 )&(data['MOS']<2.6),'MOS'] = 1
data.loc[(data['MOS']>=2.6 )&(data['MOS']<3.1),'MOS'] = 2
data.loc[(data['MOS']>=3.1 )&(data['MOS']<3.6),'MOS'] = 3
data.loc[(data['MOS']>=3.6 )&(data['MOS']<4.1),'MOS'] = 4
data.loc[(data['MOS']>=4.1 )&(data['MOS']<5),'MOS'] = 5
````

Desse modo, temos os dados distribuidos nas classes como ilustrado na célula abaixo:

````python
# Computing the mean, std, and skewness of the dataset:
unc_monovar = datahw01.describe().iloc[1:3,:-1].append(datahw01.skew()[:-1], ignore_index=True)
unc_monovar.index = ['Mean $\mu_d$','std $\sigma_d$','Skewness $\gamma_d$']
unc_monovar.to_excel('hw01-data/unc_monovar.xlsx') # Saving table
display(unc_monovar)
````

````python
# Histogram of the predictors:
fig, ax = plt.subplots(2, 4, figsize=(16, 7)) 
ax = ax.ravel()
fig.suptitle('Histogram of the Predictors')
for i in range(0, 7):
    ax[i].set_xlabel(datahw01.columns[i])
    ax[0].set_ylabel('Frequency')
    ax[4].set_ylabel('Frequency')
    ax[i].hist(datahw01.iloc[:, i], bins='sturges', alpha=.75)
fig.delaxes(ax[7])
plt.savefig('hw01-figs/hist_data.pdf')
````

### Class-conditional mono-variate analysis

````python
# Grouping each class:
cls = datahw01.groupby('MOS')
# Defining labels:
classes = ['Bad (1)', 'Poor (2)', 'Fair (3)', 'Good (4)', 'Excellent (5)']
cls_monovar = []
# Computing the mean, std, and skewness of the grouped observatios:
for i in range(len(cls.indices)):
    cls_monovar.append(cls.get_group(i+1).describe().iloc[1:3,:-1].append(cls.get_group(i+1).skew()[:-1], 
                                                                          ignore_index=True))
    cls_monovar[i].index = ['Mean $\mu_{d|l_%d}$'%(i+1),'std $\sigma_{d|l_%d}$'%(i+1),
                            'Skewness $\gamma_{d|l_%d}$'%(i+1)]
    display(classes[i], cls_monovar[i])
    cls_monovar[i].to_excel(f'hw01-data/cls_monovar{i+1}.xlsx') # Saving table
````

````python
# Histogram of the predictors for each class:
color=['C0','C1','C2','C3', 'C5']
for j in cls.indices:
    fig, ax = plt.subplots(2, 4, figsize=(16, 7)) 
    ax = ax.ravel()
    fig.suptitle(f'Histogram of the Predictors - Class: {classes[int(j)-1]}')
    for i in range(0, 8):
        ax[i].set_xlabel(cls.get_group(int(j)).columns[i])
        ax[0].set_ylabel('Frequency')
        ax[4].set_ylabel('Frequency')
        ax[i].hist(cls.get_group(j).iloc[:, i], bins='sturges', color=color[int(j)-1], alpha=.75)
    fig.delaxes(ax[7])
    plt.savefig(f'hw01-figs/hist_C{int(j)}.pdf')
````

### Tranformações no dados

Retirando a média dos dados e dividindo pelo desvio padrão. Para cada preditor $X = \{x_1,x_2,...,x_n\}$:  

$$
x_i = \frac{x_i - \mu_X}{\sigma_X}
$$

````python
# Centering and scaling the data:
scaled_datahw01 = datahw01.iloc[:,:-1]
scaled_datahw01 = (scaled_datahw01 - scaled_datahw01.mean())/scaled_datahw01.std()
display(scaled_datahw01.head())
````

### Unconditional multi-variate analysis (PCA)

Utilizando os dados transformados ```scaled_datahw01```, o passo-a-passo da análise das componentes principais 
para o conjunto de dados segue abaixo:

````python
# PCA:
# Computing the covariance matrix of the standardized data:
mt_cov = scaled_datahw01.cov().values
# Computing eigenvectors and eigenvalues:
val, vec = np.linalg.eig(mt_cov)
# Ordering the eigenvalues and the matix of the eigenvectors:
idx = val.argsort()[::-1] # Getting descending ordered indexes
val_ord = val[idx]
vec_ord = vec[:,idx]
# Defining projection matrix by selecting the first two PC's:
W = np.stack([vec_ord[:,0], vec_ord[:,1]], axis=1)
# Projecting data on the PC's:
PC = scaled_datahw01.dot(W)
# Show steps:
np.set_printoptions(4)
print(f'''Covariance matrix of the data:
{mt_cov}\n\nEigenvalues:
{val}\n\nMatix of the eigenvectors:
{vec}\n\nOrdered eigenvalues: 
{val_ord}\n\nOrdered eigenvectors:
{vec_ord}\n\nProjection matrix:
{W}\n\nTransformed dataset:
{PC.head(7)}''')
````

````python
# Scree plot:
# Computing the explained vatiance ratio:
significance = np.cumsum(((np.abs(val_ord)/val_ord.sum())))
# Plot:
fig, ax = plt.subplots()
ax.plot(np.arange(7)+1, significance, 'C2-o')
ax.plot(2,significance[1], 'o', label='Significance w/ 2 PC\'s')
ax.set_title('Explained Variance')
ax.set_xlabel('Number of PC\'s')
ax.set_ylabel('Variance')
plt.text(1.75,0.5, f'{significance[1]*100:.2f}%')
plt.legend()
plt.show()
fig.savefig('hw01-figs/expl_var.pdf')
````

````python
# PCA Plot:
cmap = ListedColormap(['orange','green','red','C3', 'c'])
fig = plt.figure(figsize=(11,7))
# Scatter plot of the PC's:
plt.scatter(PC.iloc[:,0], PC.iloc[:,1],s=35, c=datahw01['MOS'], marker='o', cmap=cmap)
# Ploting vectors:
for i in range(len(W)):
    # Arrows project features as vectors
    plt.arrow(0, 0, W[i, 0]*np.max(PC.iloc[:,0].values), 
                    W[i, 1]*np.max(PC.iloc[:,1].values),
                    color='k', width=0.04, head_width=0.3, alpha=0.65)
    # Labels
    plt.text(W[i,0]*np.max(PC.iloc[:,0])*1.2, 
             W[i, 1]*max(PC.iloc[:,1])*1.2,
             datahw01.columns.values[i], color='k', alpha=0.85)
plt.axis([ -7, 3, -5,7])
cbar = plt.colorbar()
cbar.set_label('MOS')
plt.title('PCA Biplot')
plt.xlabel('PC 0')
plt.ylabel('PC 1')
plt.show()
fig.savefig('hw01-figs/pcaBiplot.pdf')
````
